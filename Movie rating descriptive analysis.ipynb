#### Binning the data that  grouping the age data into 10 intervals so that classification becomes easy . "Binning" is part of Feature Engineering which is important in EDA.

```python
# Create buckets age-wise ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79']
label = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79']
lens["age_group"] = pd.cut(lens.age,range(0,81,10),right = False , labels=label)
lens.age_group[:5]
```

```python
# Grouping these age buckets and checking the rating behaviour. who has given max & min number of ratings
highest_rating = lens.groupby("age_group").agg({"rating" :[np.size,np.mean]})
highest_rating
```

** so the most ratings are given by from age group 20-30** but we also see ** age group 30-50 are also active in gaiving ratings** 

```python
# Get the top 100 most mentions
a = lens.groupby("movie_id").size().sort_values(ascending = False)[:100]
a
```

```python
# Setting the movie_id as the index
lens.set_index("movie_id", inplace = True)
```

### Lets check among the top 100 movies which age group user gave how much  rating to the each movie

```python
# Splitting the observations by title and age group
by_age = lens.loc[a.index].groupby(["title","age_group"]) 
#loc for slicing, groupby for arranging the table according to columns mentioned
by_age.rating.mean().head(15)
```

```python
# Use unstack on the splitted data
by_age.rating.mean().unstack().fillna(0) # unstack is used for showing the data in more understandable format
```

```python
# Reset movie_id as index
lens.reset_index("movie_id", inplace= True)
```

```python
# Pivot the data and split observations betweem male and female ratings
pivoted = lens.pivot_table(index = ["movie_id","title"],columns="sex",values="rating")
pivoted
```

```python
# Finding the rating behaviour difference between male and female
pivoted["diff"] = pivoted.M - pivoted.F
pivoted
```

```python
# Reset movie_id as index again
pivoted.reset_index("movie_id", inplace = True)
pivoted.head(5)
```

```python
# Plot the rating differences between male and female and check for ourselves for patterns
difference = pivoted[pivoted.movie_id.isin(a.index)]["diff"]
#difference.shape
difference.sort_values().plot(kind="barh",figsize=(10,20))
```

now we can see for which movie females and mens gave more ratings : values positive means men gave more rating for that particular movie and for values in negetives tells that the more ratings by women.
